| Parameters | BFS | DFS |
| --- | --- | --- |
| Stands for | BFS stands for Breadth First Search. | DFS stands for Depth First Search. |
| Data Structure | BFS(Breadth First Search) uses Queue data structure for finding the shortest path. | DFS(Depth First Search) uses <mark>Stack data structure.</mark> |
| Definition | BFS is a traversal approach in which we first walk through all nodes on the same level before moving on to the next level. | DFS is also a traversal approach in which the traverse begins at the root node and proceeds through the nodes as far as possible until we reach the node with no unvisited nearby nodes. |
| Conceptual Difference | BFS builds the tree level by level. | DFS builds the tree sub-tree by sub-tree. |
| Approach used | It works on the concept of FIFO (First In First Out). | It works on the concept of LIFO (Last In First Out). |
| Suitable for | BFS is more suitable for searching vertices closer to the given source. | DFS is more suitable when there are solutions away from source. |
| Applications | BFS is used in various applications such as bipartite graphs, shortest paths, etc. If weight of every edge is same, then BFS gives shortest path from source to every other vertex. | DFS is used in various applications such as acyclic graphs and finding strongly connected components etc. There are many applications where both BFS and DFS can be used like Topological Sorting, Cycle Detection, etc. |

In Python, arguments are

**values passed into a function when it is called**, while parameters are the variables defined in the function's signature that act as placeholders. You define arguments indirectly by specifying **parameters** in the function definition

##### 1\. Positional Arguments 

These are mapped to parameters based on their order in the function call.


```py   
    def greet(name, greeting): # name and greeting are parameters
        print(f"{greeting}, {name}!")
    
    greet("Alice", "Hello") # "Alice" and "Hello" are positional arguments
    # Output: Hello, Alice!
```


##### 2\. Keyword Arguments 

These arguments are specified using the `key=value` syntax during the function call, allowing you to <mark>pass them in any order.</mark>

```py
    def greet(name, greeting):
        print(f"{greeting}, {name}!")
    
    greet(greeting="Hi", name="Bob") # Arguments provided as key=value pairs
    # Output: Hi, Bob!
```

##### 3\. Default Arguments 

You can assign a default value to a parameter in the function definition, making the argument optional when the function is called. 

```py
    def greet(name, greeting="Hello"): # "Hello" is the default value for greeting
        print(f"{greeting}, {name}!")
    
    greet("Charlie") # Uses the default greeting
    # Output: Hello, Charlie!
    
    greet("David", "Hey") # Overrides the default value
    # Output: Hey, David!
```

##### 4\. Variable-Length Arguments 

If you do not know the number of arguments a function will need to accept, you can use arbitrary arguments.

-   **`*args` (Arbitrary Positional Arguments):** Collects a variable number of non-keyworded arguments into a tuple.
    

    ```py
        def add_numbers(*args):
            total = sum(args)
            print(f"Sum = {total}")
        
        add_numbers(1, 2, 3)    # Output: Sum = 6
        add_numbers(4, 5, 6, 7) # Output: Sum = 22
    ```
    
-   **`**kwargs` (Arbitrary Keyword Arguments):** Collects a variable number of keyword arguments into a dictionary.
    

```py
        def student_details(**kwargs):
            for key, value in kwargs.items():
                print(f"{key}: {value}")
        
        student_details(name="Eve", age=30, course="Python")
        # Output:
        # name: Eve
        # age: 30
        # course: Python
```
