# # Lab 1 codes

In Python,

**Lists**, **Arrays**, and **Dictionaries** are used to store collections of data, but they differ in structure and strictness. 

1\. List 

A **list** is a built-in, ordered collection that can store elements of **different data types**. They are highly flexible and commonly used for general data storage. 

python

    # A list with mixed data types
    my_list = [10, "Python", 3.14, True]
    
    # Accessing elements by index
    print(my_list[1])  # Output: Python
    

Use code with caution.

2\. Array 

In Python, "arrays" typically refer to **NumPy arrays** or the built-in `array` module. Unlike lists, arrays require all elements to be of the **same data type** (e.g., all integers or all floats). They are much faster and more memory-efficient for mathematical operations. 

python

    import numpy as np
    
    # An array of integers
    my_array = np.array([1, 2, 3, 4, 5])
    
    # Performing direct math operations (not possible with standard lists)
    print(my_array * 2)  # Output: [2, 4, 6, 8, 10]
    

Use code with caution.

3\. Dictionary 

A **dictionary** is a collection of **key-value pairs**. Instead of using numeric indexes (0, 1, 2...), you use unique keys (like strings or IDs) to look up information. 

python

    # A dictionary representing a user
    my_dict = {
        "name": "Alice",
        "age": 25,
        "city": "New York"
    }
    
    # Accessing values by their key
    print(my_dict["name"])  # Output: Alice
    

Use code with caution.

Quick Comparison Table 

| Feature | List | Array (NumPy) | Dictionary |
| --- | --- | --- | --- |
| **Access Method** | Index (e.g., `0`) | Index (e.g., `0`) | Key (e.g., `'name'`) |
| **Data Types** | Mixed (Heterogeneous) | Single type (Homogeneous) | Mixed (Values can be anything) |
| **Primary Use** | General collections | Math and large datasets | Labeled data/lookups |
| **Syntax** | `[1, 2, 3]` | `np.array([1, 2, 3])` | `{"key": "value"}` |

### # Lab task 1: List


````python
arr = (10,20,30,40,50)
print(arr[0:5:1])
````

````python
list_1 = [10,20,30,"CSE","BBA",[1,2,3]]
print(list_1)
print(list_1[0:5:1])
print(list_1[0:5:2])
# print(list_1.keys())
````

This code demonstrates tuple slicing using the standard syntax [start:stop:step]. 

Breakdwon of arr[0:5:1]
- arr = (10, 20, 30, 40, 50): Creates a tuple (an immutable ordered collection).
- 0 (Start): The index where the slice begins (inclusive). In Python, indexing starts at 0, so this refers to the first element, 10.
- 5 (Stop): The index where the slice ends (exclusive). The slice stops before this index. Since the tuple has 5 elements, index 5 is just past the end, so it includes everything up to the final element.
- 1 (Step): The increment between each index. A step of 1 means "take every element" without skipping. 
Result

The output will be:

(10, 20, 30, 40, 50)

Essentially, this specific slice retrieves the entire tuple exactly as it is because it starts at the beginning, goes to the end, and takes every single item. 

Shorthand Equivalent

In Python, if you want the entire sequence, you can omit these values. The following is equivalent to your code:
```
print(arr[:])
```

### # Lab task 2: dictionary


````python
my_dictionary = {'10':[10,2,5],'11':[50,60,7],'12':[1,2,3]}
print(my_dictionary)
print(my_dictionary.keys())
print(my_dictionary.values())
````

These commands are used to retrieve and display specific parts of a dictionary in Python 2026.

 

1\. `my_dictionary.keys()` 

-   **Purpose**: Returns a **view object** containing all the unique keys currently in the dictionary.
-   **Behavior**: In Python 3.x and 2026, this is a **dynamic view**, not a static list. If you add or remove keys from the dictionary, this object updates automatically to reflect those changes.
-   **Output Format**: It displays as `dict_keys(['key1', 'key2', ...])`. 

2\. `my_dictionary.values()` 

-   **Purpose**: Returns a **view object** containing all the values stored in the dictionary.
-   **Behavior**: Like the keys method, this is a dynamic view. It allows for duplicate values if they exist in the dictionary.
-   **Output Format**: It displays as `dict_values([val1, val2, ...])`. 

Key Practical Tips 

-   **Efficiency**: Because these return "views" rather than full copies of the data, they are very memory-efficient for large dictionaries.
-   **Iteration**: You can loop through these directly:
    
    python
    
        for key in my_dictionary.keys():
            print(key)
        
    
    Use code with caution.
    
-   **Converting to a List**: If you need a standard list (for example, to use an index like `[0]`), you must explicitly convert it:
    
    python
    
        list_of_keys = list(my_dictionary.keys())
        ```
        
    
    Use code with caution.
