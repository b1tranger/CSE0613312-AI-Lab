In Python, every variable refers to an

**object**, and every object is either **mutable** or **immutable** based on its type. 

Core Difference 

-   **Mutable Objects**: Can be changed after they are created. When you modify them, the object’s **identity** (memory address) remains the same.
-   **Immutable Objects**: Cannot be changed after they are created. Any operation that appears to modify them actually creates a **new object** with a new identity. 

Built-in Types Summary 

| Category | Mutable (Changeable) | Immutable (Unchangeable) |
| --- | --- | --- |
| **Collections** | `list`, `dict`, `set`, `bytearray` | `tuple`, `frozenset`, `bytes`, `range` |
| **Primitives** | _(None)_ | `int`, `float`, `bool`, `complex`, `str` |
| **Custom** | User-defined classes (default) | Classes using `__slots__` or `@dataclass(frozen=True)` |

Key Practical Implications 

-   **Memory Efficiency**: Modifying a mutable object is "cheap" because it happens in-place. Creating "modified" versions of large immutable objects (like a 1,000-item tuple) is "expensive" as it requires copying the entire structure.
-   **Dictionary Keys**: Only **immutable** objects can be used as dictionary keys or set elements. This is because they are **hashable**—their value (and thus their hash) won't change unexpectedly.
-   **Function Arguments**: When you pass a mutable object to a function, any changes made inside the function affect the original object. Immutable objects are effectively "safe" from being changed by functions in this way.
-   **Default Arguments**: Never use a mutable object (like `def func(x=[])`) as a default argument. The same list object is reused across every call, leading to unexpected data accumulation. 

The "Subtle" Exception 

A **tuple** is immutable, meaning you cannot swap its elements. However, if a tuple contains a **mutable** object (like a list), you can modify the contents of that list. The tuple’s _references_ remain fixed, but the _referenced_ objects may change.
