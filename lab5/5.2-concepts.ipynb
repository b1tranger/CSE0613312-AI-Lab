The

`if __name__ == '__main__':` statement in Python acts as a conditional guard that ensures specific code runs **only** when a script is executed directly (e.g., `python script.py`), not when it is imported as a module into another script. It acts as the program's entry point, separating executable code from reusable functions.

> python list, tuple, array, set syntax

Syntax and Examples 

| Data Type | Delimiters | Example Syntax | Key Characteristic |
| --- | --- | --- | --- |
| **List** | `[]` (square brackets) | `my_list = [1, "hello", 3.14]` | Ordered, mutable, allows duplicates. |
| **Tuple** | `()` (parentheses) | `my_tuple = (10, 20, 30)` | Ordered, **immutable**, allows duplicates. |
| **Set** | `{}` (curly braces) | `my_set = {1, 2, 3, 2}` | Unordered, mutable, **no duplicates**. |
| **Array** | `array` module | `import array` `my_array = array.array('i', [1, 2, 3])` |
